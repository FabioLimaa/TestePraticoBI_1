```python
import pandas as pd
import matplotlib.pyplot as plt
import colorsys
import seaborn as sns
import datetime

```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00381/PRSA_data_2010.1.1-2014.12.31.csv'
df = pd.read_csv(url, sep=',')
df.rename(columns={'pm2.5': 'pm25'}, inplace=True)
df.head()
```

```python
## Droping NA values
```

```python
df = df[df['pm25'].notna()]
df
```

```python
## creating timestamp for timeseries
```

```python
df['timestamp'] = df.apply(lambda x : datetime.datetime(year=x['year'], month=x['month'], day=x['day'], hour=x['hour']), axis=1)
df.head()
```

```python
##Considering the numbers from U.S. Environmental Protection Agency (2021), we can categorize the air quality as follows:
## Good  - 0 - 50 
## Moderate  - 51 - 100 
## Unhealthy for Sensitive Groups  -101 – 150
## Unhealthy  -151 – 200
## Very Unhealthy  - 201 – 300
## Hazardous  - 301 - 500


##During the given period in the data, we can figure out that:
## The mean of air quality is Moderate almost unhealth for sensitive groups
## the min value is 0 and the max is 994, neeed to check if these values can be considered as outlier
```
